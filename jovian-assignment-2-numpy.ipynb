## Function 3 - numpy.swapaxes

Interchange two axes of an array.

```python
# Example 1 - working
x = np.array([[1,2,3]])
print(x)
np.swapaxes(x,0,1) # (array , first axes , second axes)
```

Swapping columns or rows in array,  (array , first axes , second axes)

```python
# Example 2 - working
x = np.array([[[0,1],[2,3]],[[4,5],[6,7]]])
np.swapaxes(x,0,2)
```

Can work in multiple dimentions

```python
# Example 3 - breaking (to illustrate when it breaks)
np.swapaxes(x,0,5)

# axis should be in array bound
```

Arrays values in swipping should be in array shape range

Used for reorganizing columns in big arrays to deal with

## Function 4 - numpy.hsplit
Split an array into multiple sub-arrays horizontally (column-wise)

```python
# Example 1 - working
x = np.arange(16.0).reshape(4, 4)
print(x)
print()
np.hsplit(x, 2)
```

Splitting array into sub arrays

```python
# Example 2 - working
np.hsplit(x, np.array([3, 6]))
```

Horizantily splitting

```python
# Example 3 - breaking (to illustrate when it breaks)
np.hsplit(x, 5)

# split num is bigger than columns num
```
